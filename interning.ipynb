```python
x = 5
print(id(x))

y = 5
print(id(y))
print(x is y)

x = x + 1
print(id(x))

# Interning is when Python stores one shared copy of some immutable objects (strings, ints) and reuses it, instead of creating a new identical object every time.
```

```python
# Immutable = object's contents can't change (ints, strings, tuples). Operations make NEW objects, names are just rebinded to that new object
# Mutable = object's contents can change in-place (lists, dicts, sets). Methods modify the same object
```

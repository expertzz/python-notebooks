```python
from collections import Counter

words = ["house", "dog", "cat", "house"]

c = Counter(words)
c.most_common(2)

```

```python
from collections import defaultdict # auto creates missing keys

groups = defaultdict(list)

for name, team in [("Tom", "TeamA"), ("Bob", "TeamB"), ("Sob", "TeamA")]:
    groups[team].append(name)

print(groups)

###

groups_dict = {}

for name, team in [("Tom", "TeamA"), ("Bob", "TeamB"), ("Sob", "TeamA")]:
    groups_dict.setdefault(team, []).append(name)
    groups_dict[team].append(name)

print(groups_dict)
```

```python
from collections import deque

q = deque([1, 2, 3])
q.appendleft(0)
q.pop()

print(q)
```

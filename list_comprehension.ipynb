```python
grid_shallow_copy = [[0] * 3] * 3
grid_shallow_copy[0][0] = 3
print(grid_shallow_copy)
print(grid_shallow_copy[0] is grid_shallow_copy[1] is grid_shallow_copy[2])

grid_deep_copy = [[0] * 3 for _ in range(3)]
grid_deep_copy[0][0] = 3
print(grid_deep_copy)
print(grid_deep_copy[0] is grid_deep_copy[1] is grid_deep_copy[2])
```

```python
# list comprehension
# [ expression for item in iterable ]
# same as
# result = []
# for item in iterable:
#   result.append(expression)

one = [[1] * 3 for _ in range(3)]
print(one)

nums = [3,4,5,6]
strings = [str(n) for n in nums]
print(nums)

negatives = [-n for n in nums]
print(negatives)

squares = [n * n for n in nums]
print(squares)
```

```python
# transpose matrix
matrix = [
    [2,4,6],
    [4,5,6]
]

[[row[i] for row in matrix] for i in range(len(matrix[0]))]
# für jedes i für jede row, jeweils das element an Index i nehmen
```

```python
transpose = []
print(len(matrix[0]))
for i in range(len(matrix[0])):
    new_row = []
    for row in matrix:               # row is each row in the matrix
        new_row.append(row[i])       # pick the ith element from each row
    transpose.append(new_row)
print(transpose)
```

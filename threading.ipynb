```python
import threading
import requests
import time

urls = ["https://google.com"] * 10

def fetch_sequential():
    start = time.time()
    for url in urls:
        requests.get(url)
    print(f"sequential: {time.time() - start:.2f}s")

def fetch(url):
    requests.get(url)

def fetch_threaded():
    start = time.time()
    threads = [threading.Thread(target=fetch, args=(url,)) for url in urls]
    for t in threads:
        t.start()
    for t in threads:
        t.join()
    print(f"threaded: {time.time() - start:.2f}s")


fetch_sequential()
fetch_threaded()

```

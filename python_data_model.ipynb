```python
# Object Identity/Equality

class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __eq__(self, other):
        return self.x == other.x and self.y == other.y
    
    def __hash__(self):
        return hash((self.x, self.y))
    
p1 = Point(1, 2)
p2 = Point(1, 2)

print(p1 == p2) # checks value via __eq__
print(p1 is p2) # checks identity (same object in memory)
```

```python
# Container Protocol

class Deck:
    def __init__(self):
        self.cards = ["A", "K", "Q", "J"]

    def __len__(self):
        return len(self.cards)
    
    def __getitem__(self, index):
        return self.cards[index]
    
    # def __contains__(self, item):
    #     return item in self.cards
    
deck = Deck()

print(len(deck))
print(deck[1:3])
print("K" in deck)

for card in deck: # works via __getitem__
    print(card) 

# __getitem__ alone enables iteration, slicing, and in operator
```

```python
class Multiplier:
    def __init__(self, factor):
        self.factor = factor
    
    def __call__(self, x):
        return x * self.factor
    
double = Multiplier(2)
print(double(10))
print(double(20))
callable(double) # instance behaves like a function
```
